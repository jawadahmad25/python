```python
#white style
import seaborn as sns 
import matplotlib.pyplot as plt 
phool= sns.load_dataset ("iris")
print (phool)
sns.barplot (x = "species", y="sepal_width", data= phool)
#Now ploting the line graph 
plt.title (" pholo k badsha")
plt.show()
```

```python
#white style
import seaborn as sns 
import matplotlib.pyplot as plt 
phool= sns.load_dataset ("iris")
#changing size 
plt.figure(figsize=(8, 6))
#print (phool)
sns.set_style ( style=None, rc=None )
sns.set_style ("dark")
sns.barplot (x = "species", y="petal_length", data= phool)
#Now ploting the line graph 
plt.title (" pholo k badsha")
plt.grid ()
plt.show()
```

```python
#white style
import seaborn as sns 
import matplotlib.pyplot as plt 
kashti= sns.load_dataset ("titanic")
#print (kashti)
sns.barplot (x="who", y= "alone", hue= "class", data= kashti)
plt.show()
```

```python
#changing order
import seaborn as sns 
import matplotlib.pyplot as plt 
kashti= sns.load_dataset ("titanic")
#print (kashti)
sns.barplot (x="who", y= "alone", hue= "class", data= kashti, order =["woman","man", "child"])
plt.show()
```
